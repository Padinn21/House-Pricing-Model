
# Handling Missing Value

```python
# columns with most missing value
cols_nulls = df_train.isnull().sum().sort_values(ascending = False)

# make visualiazation about column with most missing value
plt.figure(figsize=(25, 10))
sns.barplot(x=cols_nulls.index, y=cols_nulls)
plt.xticks(rotation = 'vertical')
plt.show()
```

The Visualization above, give us visualization about distribution of missing value for each column and for each column with highest number of missing value, we can handle it in next process

```python
# get the exact number of missing value in each column
cols_nulls.head(20)
```

```python
# drop column with high number of missing value
df_train.drop(['PoolQC', 'MiscFeature', 'Alley'], axis = 1, inplace=True)
df_train.isnull().sum().sort_values(ascending=False)
```

```python
# make a list where's NaN value mean 'None'
none_cols = ['Fence', 'FireplaceQu', 'GarageType',
    'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType', 'Heating']

# make a list where's NaN value mean '0'
zero_cols = ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath',
    'BsmtHalfBath', 'GarageYrBlt', 'GarageArea', 'GarageCars', 'MasVnrArea']

# make a list where's NaN is actually missing and can replaced with mode
miss_cols = ['Electrical', 'Exterior1st', 'Exterior2nd', 'Functional', 'KitchenQual',
    'SaleType', 'Utilities']

# Filling the list of missing values

for col in zero_cols:
    df_train[col].replace(np.nan, 0, inplace=True)

for col in none_cols:
    df_train[col].replace(np.nan, 'None', inplace=True)

for col in miss_cols:
    df_train[col].replace(np.nan, df_train[col].mode()[0], inplace=True)

df_train['MSZoning'].replace(np.nan, 'None', inplace=True)

# filling LotFrontage according to Neighborhood
df_train['LotFrontage'] = df_train.groupby(
    ['Neighborhood'])['LotFrontage'].apply(lambda x: x.fillna(x.median()))
```

# Handling Outliers

```python
# rename column that contains number
df_train.rename(columns = {'1stFlrSF': 'OnestFlrSf' ,'2ndFlrSF': 'TwondFlrSf'}, inplace = True)
```

```python
# make function to detect Q1, Q3, and IQR
def outlier_detection(datacolumn):
  Q1,Q3 = np.percentile(datacolumn, [25,75])
  IQR = Q3 - Q1
  lowerbound = Q1 - (1.5 * IQR)
  upperbound = Q3 + (1.5 * IQR)
  return lowerbound, upperbound
```

```python
# handling outlier in OverallQual column
lowerbound, upperbound = outlier_detection(df_train.OverallQual)
df_train.drop(df_train[(df_train.OverallQual < lowerbound) | (df_train.OverallQual > upperbound)].index, inplace = True)

# handling outlier in GrLivArea column
lowerbound, upperbound = outlier_detection(df_train.GrLivArea)
df_train.drop(df_train[(df_train.GrLivArea < lowerbound) | (df_train.GrLivArea > upperbound)].index, inplace = True)

# handling outlier in GarageArea column
lowerbound, upperbound = outlier_detection(df_train.GarageArea)
df_train.drop(df_train[(df_train.GarageArea < lowerbound) | (df_train.GarageArea > upperbound)].index, inplace = True)

# handling outlier in TotalBsmtSF column
lowerbound, upperbound = outlier_detection(df_train.TotalBsmtSF)
df_train.drop(df_train[(df_train.TotalBsmtSF < lowerbound) | (df_train.TotalBsmtSF > upperbound)].index, inplace = True)

# handling outlier in LotArea column
lowerbound, upperbound = outlier_detection(df_train.LotArea)
df_train.drop(df_train[(df_train.LotArea < lowerbound) | (df_train.LotArea > upperbound)].index, inplace = True)

# handling outlier in OnestFlrSf column
lowerbound, upperbound = outlier_detection(df_train.OnestFlrSf)
df_train.drop(df_train[(df_train.OnestFlrSf < lowerbound) | (df_train.OnestFlrSf > upperbound)].index, inplace = True)

# handling outlier in SalePrice column
lowerbound, upperbound = outlier_detection(df_train.SalePrice)
df_train.drop(df_train[(df_train.SalePrice < lowerbound) | (df_train.SalePrice > upperbound)].index, inplace = True)
```

# Modelling


```python
# split x and y
y = df_train['SalePrice']
x = df_train.drop(['SalePrice'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
mms = MinMaxScaler()
x_train_scal = mms.fit_transform(x_train)
x_test_scal = mms.transform(x_test)
```

```python
linear_svr = {'epsilon': 0.1, 'max_iter': 100, 'tol': 0.001}
tree_param = {'criterion': 'squared_error', 'max_depth': 16, 'min_samples_leaf': 5, 'min_samples_split': 2, 'min_weight_fraction_leaf': 0.0}
```

```python
linreg = LinearRegression()
linearsvr = LinearSVR(**linear_svr, random_state=42, dual=False, loss='squared_epsilon_insensitive', C=1)
tree_tuned = DecisionTreeRegressor(random_state=42, **tree_param)
random_forest = RandomForestRegressor(random_state=42, **tree_param)
ridge = Ridge(max_iter=1000 ,random_state=42)
lasso = Lasso(random_state=42)
elastic = ElasticNet(random_state=42)
xgboost = XGBRegressor(random_state=42)
lgbm = LGBMRegressor(random_state=42)



model_pipeline = [linreg, tree_tuned, linearsvr, random_forest, ridge, lasso, elastic, xgboost, lgbm]
model_list = ['Linear Regression', 'Decision Tree', 'SVM', 'Random Forest', 'Ridge Regression', 'Lasso Regression', 'ElasticNet', 'xgboost', 'lgbm']
```

```python
def modelselect():
    r2score = []
    mae = []
    mse = []
    rmse = 0
    for model in model_pipeline:
      model.fit(x_train_scal, y_train)
      y_pred = model.predict(x_test_scal)
      r2score.append(metrics.r2_score(y_test, y_pred))
      mae.append(metrics.mean_absolute_error(y_test, y_pred))
      mse.append(metrics.mean_squared_error(y_test, y_pred))
      rmse = np.sqrt(mse)
    result_df = pd.DataFrame({'Model' : model_list, 'R2 Score' : r2score, 'MAE' : mae, 'MSE' : mse, 'RMSE' : rmse})
    result_df = result_df.sort_values(by='R2 Score', ascending=False)
    return result_df
```

```python
result = modelselect()
result
```
